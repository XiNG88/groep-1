# Introductie

In het boek wordt in hoofdstuk 2 geschreven over warmtetransport. Dat kan op drie manieren plaatsvinden. Het is niet eenvoudig om deze drie verschillende vormen uit elkaar te houden. In het vak 'Fysische Transportverschijnselen', dat in het tweede jaar wordt gegeven, zal je zien dat de natuurkunde achter deze verschillende vormen van warmtetransport ook best ingewikkeld is. 

```{exercise}
Welke vormen van warmtetransport bestaan er?
``````{solution}
De drie hoofdvormen van warmtetransport zijn:

- **Geleiding (conductie)**: warmte stroomt door een materiaal door botsingen/trillingen van deeltjes.
- **Convectie**: warmte wordt meegenomen door stroming van een vloeistof of gas (hier: lucht rond/in de buis).
- **Straling**: warmtetransport via elektromagnetische straling (infrarood), ook zonder medium.

Vaak spelen ze tegelijk; in deze proef zijn vooral convectie en straling belangrijk, met een kleine bijdrage van geleiding via de grijparm/ophanging.
```


In deze proef proberen we een inschatting te maken van de ordegrootte van de verschillende vormen van warmtetransport bij de koeling van een metalen buis aan lucht. 

# Theorie

Volgens [Newton's wet van afkoeling](https://nl.wikipedia.org/wiki/Afkoelingswet_van_Newton) is de snelheid waarmee een voorwerp afkoelt evenredig met het verschil in de temperatuur van het voorwerp ($T$) en de omgeving ($T_0$). We kunnen dit schrijven als:

$$
    \dot{Q} = -hA(T(t) - T_0),
$$

waarin 

- $\dot{Q}$ de warmtestroom in $\mathrm{W}$,
- $A$ het oppervlak waardoor koeling optreedt in $\mathrm{m}^2$,
- $h$ de warmteoverdrachtscoëfficiënt in $\mathrm{W/(m^2 K)}$.

Dit levert de differentiaalvergelijking

$$
    C\dot{T} = -hA(T(t) - T_0),
$$

met $C$ de warmtecapaciteit in $\mathrm{J/kg}$. Herschrijven met $\tau = \frac{C}{hA}$ levert:

$$
    -\tau\dot{T} = T(t) - T_0,
$$ (eq:dv1)

met als oplossing:

$$
    T(t) - T_0 = (T(0) - T_0)\text{e}^{-t/\tau}.
$$ (eq:dv1_opl)

We kunnen hieruit dus concluderen dat $\tau$ de karakteristieke tijdsduur is waarin de temperatuur van de buis een factor $\text{e}$ verlaagd ten opzichte van de omgevingstemperatuur. 

```{exercise} Oplossing DV controleren 
:label: ex-dvopl
Controleer bovenstaande door vergelijking [](#eq:dv1_opl) te substitueren in [](#eq:dv1).
```

We controleren of vergelijking [](#eq:dv1_opl) een oplossing is van [](#eq:dv1).

Neem
$$
T(t) = T_0 + (T(0)-T_0)\,e^{-t/\tau}.
$$
Differentieer:
$$
\dot T(t)=\frac{\mathrm dT}{\mathrm dt}= (T(0)-T_0)\left(-\frac{1}{\tau}\right)e^{-t/\tau}
= -\frac{1}{\tau}\bigl(T(t)-T_0\bigr).
$$
Vermenigvuldig met \(-\tau\):
$$
-\tau \dot T(t)= T(t)-T_0,
$$
en dat is precies [](#eq:dv1). Dus [](#eq:dv1_opl) voldoet aan de differentiaalvergelijking.


We zijn hier voor het gemak uitgegaan van een $h$ die onafhankelijk is van de temperatuur. We weten echter dat warmtetransport door straling niet lineair gaat, maar als 

$$
    \dot{Q}_s = \epsilon \sigma A (T^4 - T_0^4).
$$ 

Voor kleine temperatuurverschillen ($\Delta T = T - T_0$) is dit te vereenvoudigen tot

$$
    \dot{Q}_s = \epsilon \sigma A ((T_0+\Delta T)^4 - T_0^4) \approx \epsilon \sigma 4A T_0^3 \Delta T.
$$

Zolang $\Delta T$ dus relatief klein is ten opzichte van $T_0$, kunnen we $h$ dus inderdaad als een constante beschouwen. 

Laat zien dat bovenstaande geldt door eerst $(T_0+\Delta T)^4$ uit te schrijven, te substitueren en dan te bedenken dat $\Delta T$ klein is en $\Delta T^2$ dus nog kleiner.

Werk de binomiale expansie uit:
$$
(T_0+\Delta T)^4
= T_0^4 + 4T_0^3\Delta T + 6T_0^2(\Delta T)^2 + 4T_0(\Delta T)^3 + (\Delta T)^4.
$$
Daarom is
$$
(T_0+\Delta T)^4 - T_0^4
= 4T_0^3\Delta T + 6T_0^2(\Delta T)^2 + 4T_0(\Delta T)^3 + (\Delta T)^4.
$$
Als \(\Delta T\) klein is t.o.v. \(T_0\), dan zijn \((\Delta T)^2,(\Delta T)^3,(\Delta T)^4\) veel kleiner dan \(\Delta T\) en mogen die hogere-orde-termen worden verwaarloosd. Dan blijft over:
$$
(T_0+\Delta T)^4 - T_0^4 \approx 4T_0^3\Delta T.
$$



Hoe groot is de fout in het warmtetransport door straling die we maken voor het temperatuurbereik waarin we deze proef uitvoeren?

De *exacte* stralings-term is
$$
\dot Q_s \propto (T^4-T_0^4),
$$
en de lineaire benadering is
$$
\dot Q_{s,\text{lin}} \propto 4T_0^3(T-T_0).
$$
De relatieve fout (alle constanten vallen weg) is dus:
$$
\varepsilon_\text{rel}(T)=\frac{\dot Q_{s,\text{lin}}-\dot Q_s}{\dot Q_s}
= \frac{4T_0^3(T-T_0)-(T^4-T_0^4)}{T^4-T_0^4}.
$$

Neem als omgevingstemperatuur $$T_0\approx 20^\circ\text{C}=293{,}15\,\text{K}$$ en een buistemperatuur in de proef van ongeveer $$45{-}60^\circ\text{C}$$ ($$318{-}333\,\text{K}$$). Dan krijg je (afgerond):

- bij $$T=45^\circ\text{C}$$: $$\varepsilon_\text{rel}\approx -12\%$$
- bij $$T=60^\circ\text{C}$$: $$\varepsilon_\text{rel}\approx -18\%$$

Dus in dit temperatuurbereik onderschat de linearisatie het warmtetransport door straling met grofweg **10–20%**.  
Let op: dit is de fout op *alleen* de stralingsterm; de fout op de totale warmtestroom is kleiner als convectie ook een groot deel bijdraagt.

**Buitenoppervlak (externe oppervlakte)**  
Voor een (bij benadering) cilindrische buis met buitendiameter \(D\) en lengte \(L\):

$$
A_\text{buiten}\approx \pi D L + 2\pi\left(\frac{D}{2}\right)^2,
$$
(laterale mantel + twee kopse vlakken). In veel gevallen is het kopse-oppervlak klein t.o.v. \(\pi DL\) en mag je die term eventueel verwaarlozen.

**Warmtecapaciteit van de buis**  
$$
C_\text{buis} = m\,c.
$$
Meet \(m\) (massa) direct, of bepaal \(m\) uit de geometrie en de dichtheid \(\rho\). Voor een dunwandige buis met wanddikte \(t\):

$$
V_\text{metaal}=\pi\left(r_o^2-r_i^2\right)L,\quad r_o=\frac{D}{2},\quad r_i=r_o-t,
$$
$$
m=\rho V_\text{metaal}.
$$

**Omgevingstemperatuur**  
Dit is de luchttemperatuur in het lokaal (meet met een thermometer). In de analyse hieronder heb ik als voorbeeld $T_0 = 20^\circ\text{C}$ gebruikt; vervang dit door jullie gemeten waarde.

:label: ex_conf
Maakt het uit in welke richting je de (cilindrische) as van de buis legt? En waarom zit er een dop bij de buis?
```

```{solution} ex_conf
Ja, dit kan uitmaken, vooral door **natuurlijke convectie**.

- **Verticale buis zonder dop (open):** warme lucht in de buis wordt lichter en stijgt op. Er ontstaat een “schoorsteen‑effect” (lucht stroomt onderin naar binnen, warmt op, stijgt en verlaat de buis bovenin). Dat verhoogt de convectieve warmtestroom en dus verlaagt \(\tau\) (sneller afkoelen).
- **Met dop:** de lucht in de buis kan minder makkelijk worden ververst, waardoor interne convectie veel kleiner wordt. De buis koelt dan relatief langzamer.
- **Horizontaal:** het schoorsteen‑effect in de buis is (bijna) weg. Ook de stroming rond de buis kan anders zijn, waardoor \(h\) en/of de effectieve bijdrage van interne convectie verandert.

Kortom: de **oriëntatie** beïnvloedt de natuurlijke convectie (rond én in de buis), en de **dop** is bedoeld om de interne luchtstroming (en daarmee extra convectie) te onderdrukken.
```


Doe dit voor twee of drie configuraties:

1. De buis met de as in verticale richting en afgesloten met dop.
2. De buis met de as in verticale richting *zonder* dop.
3. (alleen bij voldoende tijd) De buis met de as in horizontale richting en afgesloten met dop.

```{warning} Veiligheid
Zorg ervoor dat je de buis niet aanraakt terwijl deze warm is. De metalen buis geleidt warmte. Zelfs een temperatuur van 60 graden celsius is daarom al onaangenaam hoog. Hou de pan met het warme water ook op een veilige plek zodat deze niet kan omkiepen en het warme water over de grond opspat.
```

# Resultaten

```````python
# Hier de data en de analyse

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from scipy.optimize import curve_fit

# 1) Data inlezen (Excel)
excel_path = Path("buizenproef.xlsx")
if not excel_path.exists():
    excel_path = Path("/mnt/data/buizenproef.xlsx")  # werkt in de sandbox

df = pd.read_excel(excel_path, sheet_name=0, header=None)

# Temperatuurkolommen (zie het Excel-bestand):
# kolom 0  = verticaal zonder dop
# kolom 3  = verticaal met dop
# kolom 7  = horizontaal zonder dop
# kolom 10 = horizontaal met dop
series_cols = {
    "Verticaal zonder dop": 0,
    "Verticaal met dop": 3,
    "Horizontaal zonder dop": 7,
    "Horizontaal met dop": 10,
}

temps_series = {}
for name, col in series_cols.items():
    temps = df[col].dropna().astype(float).to_numpy()
    temps_series[name] = temps

dt = 5.0  # seconden tussen metingen

# 2) Constantes (vul aan met jullie metingen!)
T_omg = 20.0  # °C, omgevingstemperatuur (meet & vervang indien mogelijk)

# Geometrie buis (voorbeeldwaarden; vervang door jullie gemeten waarden)
D = 0.022   # m, buitendiameter
L = 0.20    # m, lengte
t = 0.002   # m, wanddikte

# Materiaalkeuze (voorbeeld: aluminium; vervang indien nodig)
rho = 2700.0   # kg/m^3
c = 880.0      # J/(kg K)

r_o = D/2
r_i = max(r_o - t, 0.0)

# Buitenoppervlak (mantel + 2 kopse vlakken)
buitenoppervlak = np.pi * D * L + 2 * np.pi * r_o**2  # m^2

# Massa en warmtecapaciteit (J/K)
V_metaal = np.pi * (r_o**2 - r_i**2) * L
m_buis = rho * V_metaal
warmtecapaciteit = m_buis * c

print(f"Buitenoppervlak A = {buitenoppervlak:.5f} m^2")
print(f"Massa buis m = {m_buis:.4f} kg")
print(f"Warmtecapaciteit C = {warmtecapaciteit:.2f} J/K")
print(f"Gebruikte omgevingstemperatuur T_omg = {T_omg:.1f} °C")

# 3) Fit: Newton afkoeling met vaste T_omg
def exp_func_fixed(t, A, tau):
    # A = (T(0)-T_omg) bij t=0 (in °C)
    # tau = karakteristieke tijd (s)
    return A * np.exp(-t / tau) + T_omg

fit_results = []

for name, temps in temps_series.items():
    times = np.arange(len(temps)) * dt

    # Startschatting
    p0 = [temps[0] - T_omg, 400.0]
    bounds = ([0.0, 1.0], [200.0, 5000.0])

    popt, pcov = curve_fit(exp_func_fixed, times, temps, p0=p0, bounds=bounds, maxfev=20000)
    A_exp, tau_exp = popt

    # Warmteoverdrachtscoëfficiënt
    h_exp = warmtecapaciteit / (tau_exp * buitenoppervlak)  # W/(m^2 K)

    # Goodness-of-fit (R^2)
    residuals = temps - exp_func_fixed(times, *popt)
    ss_res = float(np.sum(residuals**2))
    ss_tot = float(np.sum((temps - np.mean(temps))**2))
    r2 = 1.0 - ss_res/ss_tot if ss_tot > 0 else np.nan

    fit_results.append({
        "Configuratie": name,
        "A = T(0)-T_omg (°C)": A_exp,
        "tau (s)": tau_exp,
        "h (W/m^2K)": h_exp,
        "R^2": r2,
        "N": len(temps),
    })

    # Plot per configuratie
    plt.figure()
    plt.plot(times, temps, 'o', label="meting")
    y_fit = exp_func_fixed(times, *popt)
    plt.plot(times, y_fit, '-', label=fr"fit: $\tau$={tau_exp:.0f} s")
    plt.xlabel("tijd (s)")
    plt.ylabel("temperatuur (°C)")
    plt.title(name)
    plt.grid(True)
    plt.legend()
    plt.show()

results_df = pd.DataFrame(fit_results).set_index("Configuratie")
display(results_df)

# 4) (Optioneel) indicatie straling vs convectie
sigma = 5.670374419e-8  # W/m^2/K^4
eps = 0.8               # emissiviteit (ruwe schatting; 0.2-0.9 afhankelijk van oppervlak)
T0_K = T_omg + 273.15
h_rad = 4 * eps * sigma * T0_K**3

print(f"Schatting h_straling ≈ {h_rad:.2f} W/m^2K (met eps={eps})")
print("Vergelijk dit met h uit de fits om te schatten hoeveel straling bijdraagt.")
```````

# Discussie en conclusie

## Belangrijkste resultaten (op basis van de fit)

We hebben de afkoelcurve van de buis in vier configuraties gefit met Newtons wet van afkoeling:

$$
T(t) = (T(0)-T_0)\,e^{-t/\tau} + T_0.
$$

Uit de fit volgt voor elke configuratie een karakteristieke tijd \(\tau\) en (gegeven \(A\) en \(C\)) een warmteoverdrachtscoëfficiënt
$$
h = \frac{C}{\tau A}.
$$

In onze metingen liggen de \(\tau\)-waarden in de orde van enkele honderden seconden. De verschillen tussen configuraties zijn niet extreem groot, maar wel zichtbaar.

## Interpretatie

- **Oriëntatie en dop** beïnvloeden vooral de **natuurlijke convectie** rond de buis en (bij open buizen) ook **in** de buis.
- Een **open verticale buis** kan extra convectie krijgen door een schoorsteen‑effect (luchtverversing door opstijgende warme lucht).  
- Een **dop** onderdrukt die interne stroming en kan de effectieve warmtestroom verlagen (dus grotere \(\tau\)).  
- In de **horizontale** stand is het schoorsteen‑effect vrijwel afwezig; verschillen worden dan vooral bepaald door convectie rondom de buis en straling.

## Onzekerheden / verbeterpunten

- De fit is gevoelig voor de gekozen **omgevingstemperatuur \(T_0\)** wanneer je meetreeks niet lang genoeg doorloopt om dicht bij \(T_0\) te komen. Het is daarom belangrijk om \(T_0\) apart te meten (of lang genoeg door te meten).
- De berekende \(h\) hangt lineair af van de ingevoerde **geometrie** en **warmtecapaciteit**. Kleine fouten in diameter, lengte, wanddikte of materiaalkeuze geven direct een merkbare fout in \(h\).
- Straling kan in dit temperatuurbereik een bijdrage leveren van enkele \(\mathrm{W/(m^2K)}\) (afhankelijk van emissiviteit), dus is niet altijd te verwaarlozen.

## Conclusie

De afkoeling van de buis kan goed worden beschreven met Newtons wet van afkoeling (exponentieel verval). De oriëntatie en het wel/niet gebruiken van een dop beïnvloeden de effectieve warmteoverdracht vooral via veranderingen in natuurlijke convectie (en in mindere mate via straling).
